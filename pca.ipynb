## Principle Component Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

This notebook concerns with applying principle component analysis on our standardised dataset. We will use the scree plot to determine the number of components to use for feature selection.

```python
from sklearn.decomposition import PCA
```

```python
X_train = pd.read_csv('data/input/standardized_X_train.csv', index_col=0)
y_train = pd.read_csv('data/input/y_log_train.csv', index_col=0)
```

```python
pca = PCA()
pcscores_full = pca.fit_transform(X_train)
```

```python
pve = pca.explained_variance_ratio_
```

```python
plt.figure(figsize=(12, 6))
plt.plot(np.arange(len(pve)) + 1, np.cumsum(pve))
plt.xlabel('Components Used')
plt.ylabel('PVE')
plt.show()
```

To determine the bend, we find the iterate the cummulative sum of the explained variance ratio until the sum is greater than 0.9999. It means that the all components before this point explain 99.99% of the variance (information) in the data.

```python
np.nonzero(np.cumsum(pve) > 0.9999)
```

Therefore, we can use first 120 components generally without any loss of information, compared to all 165 features.
